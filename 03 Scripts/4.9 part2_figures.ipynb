## Histogram for order time

```python
# Create histogram
complete_df['order_time'].hist()

# Add axis labels
plt.xlabel('Order Time')     # X-axis label
plt.ylabel('Frequency')      # Y-axis label

# Add a title
plt.title('Distribution of Order Times')

# Save the figure as a PNG file
plt.savefig(r'C:\Users\analy\Documents\Instacart Basket Analysis\04 Analysis\Visualization\Distribution of Order Times.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# Save the figure as a PNG file
plt.savefig(r'C:\Users\analy\Documents\Instacart Basket Analysis\04 Analysis\Visualization\Distribution of Order Times.png', dpi=300, bbox_inches='tight')
```

The histogram shows that the ordering starts around 7-8 in the morning with the peak between 14:00-16:00. the morning hours are more busy than the afternoon.

## Number of Orders distribution among Loyality_flag


```python
# Count orders per Loyality_flag
loyality_orders = complete_df.groupby('loyalty_flag')['order_number'].count()

# Create and customize the plot
orders_loy= loyality_orders.plot(kind='bar', color='blue')
orders_loy.set_xlabel('Loyality Flag')
orders_loy.set_ylabel('Number of Orders')
orders_loy.set_title('Orders by Loyality Flag')

# Save the figure
barfig = orders_loy.get_figure()
barfig.savefig(r'C:\Users\analy\Documents\Instacart Basket Analysis\04 Analysis\Visualization\Orders by Loyality Flag.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

The regular customers have the highest number of orders

## Family Status by Age
