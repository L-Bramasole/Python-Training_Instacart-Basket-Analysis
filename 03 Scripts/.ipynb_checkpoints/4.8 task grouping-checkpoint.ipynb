#### spending habits based on customer Loyality

```python
pro_or_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

the max number 99999.0 looks like not realistic number. therefore, i will check how many rows contain this value as follows:

```python
pro_or_merge[pro_or_merge['prices'] == 99999.0]['loyalty_flag'].value_counts()
```

```python
pro_or_merge['prices'] = pro_or_merge['prices'].replace(99999.0, np.nan)
```

```python
pro_or_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

looks like the price 14900 is an outlier shifting the mean above. to make sure that this is a true price, i would check what product is that and whether it makes sense.

```python
pro_or_merge[pro_or_merge['prices'] == 14900][['product_name', 'prices']]
```

```python
pro_or_merge[pro_or_merge['prices'] == 14900]['product_name'].unique()
```

according to the check up above about the price 14900, it apears that it shows up under only one product name 'Lowfat 2% Milkfat Cottage Cheese'. which should have the price 4.90 according to instacart website. Therefore, I will replace it in the dataframe as follows

```python
pro_or_merge['prices'] = pro_or_merge['prices'].replace(14900, 4.90)
```

```python
pro_or_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

Answer: looks like there are no differences in product prices between the loyaliy groups. and the prices has normal distribution where the mean and the median are close.

#### spending flag based on mean_order

```python
# creating order_mean column grouped by user_id
pro_or_merge['mean_order'] = pro_or_merge.groupby(['user_id'])['order_number'].transform(np.mean)
```

```python
pro_or_merge.head()
```

```python
# creating spending flag column
pro_or_merge.loc[pro_or_merge['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
pro_or_merge.loc[pro_or_merge['mean_order']<10, 'spending_flag'] = 'Low spender'
```

```python
pro_or_merge['spending_flag'].value_counts(dropna= False)
```

big difference in spending based on our criteria

#### order frequency flag

```python
# creating last_order_median column grouped by days_since_prior_order
pro_or_merge['last_order_median'] = pro_or_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
pro_or_merge.head()
```

```python
# creating frequency flag column based on criterias
pro_or_merge.loc[pro_or_merge['last_order_median']>20, 'frecuency_flag'] = 'Non-frequent customer'
```

```python
pro_or_merge.loc[(pro_or_merge['last_order_median']>10) & (pro_or_merge['last_order_median']<=20), 'frecuency_flag'] = 'Regular customer'
```

```python
pro_or_merge.loc[pro_or_merge['last_order_median']<=10, 'frecuency_flag'] = 'frequent customer'
```
