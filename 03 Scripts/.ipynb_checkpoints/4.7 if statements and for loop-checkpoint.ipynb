#### if statement example

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Midume'
  elif row['prices'] > 15:
    return 'High'
  else: return 'Not enough data'
```

```python
sub_pr_o['price_range']= sub_pr_o.apply(price_label,axis=1)
```

```python
sub_pr_o['price_range'] = sub_pr_o['price_range'].astype('category')
```

# busiest hours of the day

```python
pr_or_merged["order_time"].value_counts()
```

```python
order_counts = pr_or_merged['order_time'].value_counts()
```

#### i catagorized the order number based on my estimation as follows

```python
ordertimes = []

for hour in pr_or_merged['order_time']:
    count = order_counts.get(hour, np.nan)  # Get the count for that hour
    if np.isnan(count):
        ordertimes.append(np.nan)  # Assign NaN directly
    if count > 2500000:
        ordertimes.append('Most orders')
    elif count < 110000:
        ordertimes.append('Fewest orders')
    else:
       ordertimes.append('Average orders')
```

```python
pr_or_merged['busiest_period_of_day'] = ordertimes
```

```python
pr_or_merged['busiest_period_of_day'].value_counts(dropna= False)
```
